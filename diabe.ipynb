```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Glucose'], kde=True, bins=30)
plt.title('Distribution of Glucose')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title('Glucose vs Outcome')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['BloodPressure'], kde=True, bins=30)
plt.title('Distribution of Blood Pressure')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='BMI', hue='Outcome', data=df)
plt.title('Age vs BMI by Outcome')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['Non-Diabetic', 'Diabetic']
sizes = df['Outcome'].value_counts().values
colors = ['#66b3ff','#ff9999']

plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Diabetes Cases')
plt.axis('equal')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.countplot(x='Pregnancies', data=df, palette='Set2')
plt.title('Count of Pregnancies')
plt.xlabel('Pregnancies')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(x='Outcome', y='Glucose', data=df, palette='pastel')
plt.title('Average Glucose by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Average Glucose')
plt.xticks([0,1], ['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

x = df['Glucose']
y = df['BMI']
z = df['Age']
c = df['Outcome']

scatter = ax.scatter(x, y, z, c=c, cmap='coolwarm', s=50)
ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
plt.title('3D Scatter Plot of Glucose, BMI, and Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df_grouped = df.groupby('Outcome').mean().T

plt.figure(figsize=(10,5))
for outcome in df_grouped.columns:
    plt.plot(df_grouped.index, df_grouped[outcome], label=f'Outcome {outcome}')
    
plt.title('Average Feature Values by Outcome')
plt.xlabel('Features')
plt.ylabel('Average Value')
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
